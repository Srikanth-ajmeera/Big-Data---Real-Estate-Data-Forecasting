```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# List of cities to analyze
cities = ["New York", "Los Angeles", "Chicago", "Houston"]

# Create a subplot grid
fig, axes = plt.subplots(len(cities), 4, figsize=(20, 16))
fig.suptitle("Time Series Decomposition of Housing Prices", fontsize=18, y=1.02)

for i, city in enumerate(cities):
    # Extract the time series for the city
    ts = df[df["RegionName"] == city].iloc[0, 8:]
    ts.index = pd.to_datetime(ts.index)
    ts = ts.astype(float)

    # Perform decomposition
    result = seasonal_decompose(ts, model='additive', period=12)

    # Plot original, trend, seasonal, and residual components
    axes[i, 0].plot(ts)
    axes[i, 0].set_title(f"{city} - Observed")

    axes[i, 1].plot(result.trend)
    axes[i, 1].set_title(f"{city} - Trend")

    axes[i, 2].plot(result.seasonal)
    axes[i, 2].set_title(f"{city} - Seasonal")

    axes[i, 3].plot(result.resid)
    axes[i, 3].set_title(f"{city} - Residual")

# Improve layout
for ax in axes.flatten():
    ax.tick_params(axis='x', labelrotation=45)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Load and preprocess data
df1 = df.copy()
df1.set_index('RegionName', inplace=True)

# Select top 3 cities
cities = ['New York', 'Los Angeles', 'Chicago']
forecast_months = 12

# Store results
results = {}

for city in cities:
    series = df1.loc[city, '2000-01-31':].T
    series.index = pd.to_datetime(series.index)
    series = series.astype(float)

    # Train/Test split
    train = series[:-forecast_months]
    test = series[-forecast_months:]

    ### 1. ARIMA
    arima_model = ARIMA(train, order=(2, 1, 2)).fit()
    arima_forecast = arima_model.forecast(steps=forecast_months)
    arima_mse = mean_squared_error(test, arima_forecast)

    ### 2. Linear Regression
    X = np.arange(len(series)).reshape(-1, 1)
    y = series.values.reshape(-1, 1)
    X_train, X_test = X[:-forecast_months], X[-forecast_months:]
    y_train, y_test = y[:-forecast_months], y[-forecast_months:]

    linreg = LinearRegression().fit(X_train, y_train)
    linreg_pred = linreg.predict(X_test)
    linreg_mse = mean_squared_error(y_test, linreg_pred)

    ### 3. Random Forest
    rf = RandomForestRegressor(n_estimators=100, random_state=42)
    rf.fit(X_train, y_train.ravel())
    rf_pred = rf.predict(X_test)
    rf_mse = mean_squared_error(y_test, rf_pred)

    ### Store
    results[city] = {
        'actual': test,
        'ARIMA': arima_forecast,
        'Linear Regression': linreg_pred,
        'Random Forest': rf_pred,
        'mse': {
            'ARIMA': arima_mse,
            'Linear Regression': linreg_mse,
            'Random Forest': rf_mse
        }
    }

# Plotting
for city in cities:
    plt.figure(figsize=(12, 5))
    actual = results[city]['actual']
    plt.plot(actual.index, actual.values, label='Actual', color='black', linewidth=2)
    plt.plot(actual.index, results[city]['ARIMA'], label='ARIMA')
    plt.plot(actual.index, results[city]['Linear Regression'], label='Linear Regression')
    plt.plot(actual.index, results[city]['Random Forest'], label='Random Forest')

    plt.title(f"Forecast Comparison for {city}")
    plt.ylabel("Price")
    plt.legend()
    plt.grid()
    plt.show()

    print(f"🔍 MSE for {city}:")
    for model, mse in results[city]['mse'].items():
        print(f"{model}: {mse:.2f}")
    print('-' * 60)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Load and preprocess data
df1 = df.copy()
df1.set_index('RegionName', inplace=True)

# Select top 3 cities
cities = ['New York', 'Los Angeles', 'Chicago']
forecast_months = 12

# Store results
results = {}

for city in cities:
    series = df1.loc[city, '2000-01-31':].T
    series.index = pd.to_datetime(series.index)
    series = series.astype(float)

    # Train/Test split
    train = series[:-forecast_months]
    test = series[-forecast_months:]

    ### 1. ARIMA
    # Try a different order or a different model like SARIMA if the issue persists
    try:
        arima_model = ARIMA(train, order=(2, 1, 2)).fit()
    except np.linalg.LinAlgError:
        print(f"ARIMA model failed to converge for {city}. Trying a simpler model...")
        try:
            arima_model = ARIMA(train, order=(1, 1, 1)).fit()  # Simplified order
        except np.linalg.LinAlgError:
            print(f"ARIMA model failed for {city}. Skipping...")
            continue  # Skip to the next city if still fails

    arima_forecast = arima_model.forecast(steps=forecast_months)
    arima_mse = mean_squared_error(test, arima_forecast)

    ### 2. Linear Regression
    X = np.arange(len(series)).reshape(-1, 1)
    y = series.values.reshape(-1, 1)
    X_train, X_test = X[:-forecast_months], X[-forecast_months:]
    y_train, y_test = y[:-forecast_months], y[-forecast_months:]

    linreg = LinearRegression().fit(X_train, y_train)
    linreg_pred = linreg.predict(X_test)
    linreg_mse = mean_squared_error(y_test, linreg_pred)

    ### 3. Random Forest
    rf = RandomForestRegressor(n_estimators=100, random_state=42)
    rf.fit(X_train, y_train.ravel())
    rf_pred = rf.predict(X_test)
    rf_mse = mean_squared_error(y_test, rf_pred)

    ### Store
    results[city] = {
        'actual': test,
        'ARIMA': arima_forecast,
        'Linear Regression': linreg_pred,
        'Random Forest': rf_pred,
        'mse': {
            'ARIMA': arima_mse,
            'Linear Regression': linreg_mse,
            'Random Forest': rf_mse
        }
    }
```

```python
for city in cities:
    plt.figure(figsize=(12, 5))
    actual = results[city]['actual']
    plt.plot(actual.index, actual.values, label='Actual', color='black', linewidth=2)
    plt.plot(actual.index, results[city]['ARIMA'], label='ARIMA')
    plt.plot(actual.index, results[city]['Linear Regression'], label='Linear Regression')
    plt.plot(actual.index, results[city]['Random Forest'], label='Random Forest')

    plt.title(f"Forecast Comparison for {city}")
    plt.ylabel("Price")
    plt.legend()
    plt.grid()
    plt.show()

    print(f"🔍 MSE for {city}:")
    for model, mse in results[city]['mse'].items():
        print(f"{model}: {mse:.2f}")
    print('-' * 60)
```
